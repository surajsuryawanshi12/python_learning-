```python
# nested if-else statement are if else structure inside another if else
# this allows you to check for ore complex conditions.

# example:-
num = eval(input(" enter the number "))
if num >=0 :
  if num == 0: # false
    print("zero")
  else:
    print("positive number")
else:
  print("negative number")

  # explanation if num is greater than or equal to 0, another if checks if num is 0,
  # if true, 'zero' is printed. Otherwise, 'Positive Number" is printed.
  # if num is less than 0, 'Negative Number' is printed.


```

```python
# elif format

num = eval(input(" enter the number "))

if num > 0:
  print("positive number")
elif num ==0:
  print("zero number")
else:
  print("negative number")
```
