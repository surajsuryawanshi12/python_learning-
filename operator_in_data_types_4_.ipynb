```python
# set operator
# & : Intersection ( common elements)
# | : union all elements
# -- : diffrence ( all the elements of one set except the common
# ^ : set diffrence ( all the elements form both set except the common) ) [ union - intersection]

_set ={1,2,3,4}
_setb = {4,5,6}
print(_set & _setb)
print(_set | _setb)
print(_set - _setb) # elements that only presennt in set1 not in set 2  ignore common

print(_set ^ _setb)
```
