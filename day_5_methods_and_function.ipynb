```python
# zip() ---> zips the data : Combine two values present on same index
_list = [ 1,2,3,4,5,6]
_list2 = [11,12,13,14,15,16]
_list3 = [21,22,23,24,25,26]

_listx = zip(_list, _list2, _list3)
print(list(_listx))
print(list(zip(_list, _list2, _list3)))
```
