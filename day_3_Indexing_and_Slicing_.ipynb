# `Slicing` on list, string & tuple

````python
# list are finding using three processs [ start , stop, step]
````

````python
_list =[5,6,7,8,9] # Full Data
# [6,7,8] --> SLICE
# _ LIST[start: stop]
# _list[i : j]
# start/ i : first element in slice, what is the index of that element in full data
# i : 1
# stop/ j : Last element in slice, what is the index of that element in full data
# j:  3
print(_list[1:3]) # 3 is exclusive
print(_list[1:4]) # 4 is exclusive ans is  [6,7,8]
````

````python
_list = [5,6,7,8,9] # [5,6]
_list[0:2] # 2 is exclusive
print(_list[-5:-3]) # negative
print(_list[0:2]) # positiive
````

````python
_list = ['a','b','c','d','e','f']
# ['c','d','e','f'] : slice
# _i = 2
# _j = 5+1
print(_list[2:6]) # 5 index is exclusive +1 @ positive
# for ngegative  niche
# your j is blank

print(_list[-4:])
````

````python
# edge case
# i as a blank | Start slicing from the first element | when your fist element in slice is same as first element in main data
# J as a blank | keep slicing till the end |  when your last element of slice is same as last element in the main data
````

````python
# slice --> ['a','b','c']
print(_list[:-3])
print(_list[0:3])
````

````python
_list = ['a','b','c','d','e','f'] # slice [d,e,f]
#positive
# i:3
# j: blank, 5 +1  # exclusive

print(_list[3:])
print(_list[3:6])

# negative
print(_list[-3:])
````

````python
# [start, stop, step]
````

````python
# Step
_list = ['a','b','c','d','e','f','g']
# slice = ['b','c','d','e','f']
print(_list[1:-1:1])

# k= 1 , here k: step or jump or skip it is determine the forword direction and reverse direction
# IF K + THE IT GOES TO FORWORD
# IF K - THEN IT GOES TO REVERSE

print(_list[1:6])
# k = 2
print(_list[1:6:2])
print(_list[1: : 1 ])
print(_list[0:5:2])
````

````python
# IF K + THE IT GOES TO FORWORD
# IF K - THEN IT GOES TO REVERSE
````

````python
_list = [5,6,7,8,9]
# [8,7,6] ---> Reverse slicing
print(_list[-2: -5:-1])
print(_list[-5: -2:1])
````

````python
_string = 'hello world' # slice "dlro"
print(_string.index(' '))
print(_string[-1:-5:-1])
print(_string[:6:-1])
print(_string[10:6:-1])
````
